```python
import numpy as np
from numba import autojit
import matplotlib.pyplot as plt
```

```python
theta = 4
def mandelbrot(Re, Im, max_iter):
    C = complex(Re, Im)
    Z = 0.0j
    
    for i in range(max_iter):
        Z = Z*Z + C
        if(Z.real*Z.real + Z.imag*Z.imag) >= theta:
            return i
        
    return max_iter
```

```python
columns = 2000
rows = 2000

result = np.zeros([rows,columns])
for row_index, Re in enumerate(np.linspace(-2, 1, num = rows)):
    for columns_index, Im in enumerate(np.linspace(-1, 1, num = columns)):
        result[row_index,columns_index] = mandelbrot(Re, Im, 200)
```

```python
plt.figure(dpi = 500)
plt.imshow(result.T, interpolation = "bilinear", extent = [-2,1,-1,1])
plt.savefig("mandel_4", dpi = 1000, quality = 95)
plt.xlabel("Re")
plt.ylabel("IM")
plt.xticks(())
plt.yticks(())
plt.show()
```
